```python
import matplotlib.pyplot as plt

X = [1, 2, 3, 4, 5]
Y1 = [0.77, 0.92, 0.95, 0.97, 0.97]
Y2 = [0.92, 0.94, 0.93, 0.97, 0.95]

plt.figure(figsize=(10, 6))
plt.plot(X, Y1, marker='o', label='Train', color='blue')
plt.plot(X, Y2, marker='o', label='Validation', color='red')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('IndoBERT')
plt.legend()
plt.grid(True)
plt.savefig('IndoBERT_plot.jpg', format='jpg')
plt.show()
```
